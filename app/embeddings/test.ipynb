```python
from langchain_core.documents import Document
from typing import List
import re
file_path = DATA_DIR/"raw"/"customer_policy.txt"
def load_policy_as_documents(file_path: str) -> List[Document]:
    """
    Đọc file txt chính sách và trả về list[Document].
    Mỗi điều khoản POL-XX thành 1 Document.
    """
    documents = []
    with open(file_path, "r", encoding="utf-8") as f:
        content = f.read()

    # Tách theo POL-XX:
    # pattern: 'POL-' + 2 chữ số + ':'  
    clauses = re.split(r"(POL-\d{2}:)", content)

    for i in range(1, len(clauses), 2):
        code = clauses[i].strip()       # POL-01, POL-02,...
        text = clauses[i+1].strip()     # nội dung điều khoản
        full_text = f"{code} {text}"    # ghép lại thành 1 điều khoản hoàn chỉnh

        documents.append(
            Document(
                page_content=full_text,
                metadata={
                    "code": code,
                    "source": file_path
                }
            )
        )

    return documents

# ---------------------------
# Example sử dụng
docs = load_policy_as_documents(file_path)
print(type(docs))
print(type(docs[0]))
```
